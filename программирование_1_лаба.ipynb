```python
# Задание task_02_15.
# Выполнил: Манидарова Н.Ч.
# Группа: ЦИБ-251

p = int(input("Грузоподъемность грузовика (кг.) = "))
n = int(input("Количество предметов = "))
total = 0
for i in range(1, n + 1):
    mass = int(input(f"Масса {i}-го предмета (кг.) = "))
    total += mass

if total <= p:
    print("Да")
else:
    print("Нет")
```
