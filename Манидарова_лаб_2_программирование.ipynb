```python
# Задание task_02_23.
# Выполнил: Манидарова Н.Ч.
# Группа: ЦИБ-251

#Дано натуральное число n. Определите, является ли оно членом последовательности Фибоначчи.
import math

n = int(input("n = "))

def is_fibonacci(num):
    if num < 0:
        return False
    if num == 0 or num == 1:
        return True

    def is_perfect_square(x):
        if x < 0:
            return False
        root = int(math.isqrt(x))
        return root * root == x

    return is_perfect_square(5 * num * num + 4) or is_perfect_square(5 * num * num - 4)

print("Является" if is_fibonacci(n) else "Не является")
```

```python
# Задание task_02_27.
# Выполнил: Манидарова Н.Ч.
# Группа: ЦИБ-251

#Выведите на экран (в строку) n первых простых чисел.


def print_first_n_primes(n):
    if n <= 0:
        return

    primes = []
    candidate = 2

    while len(primes) < n:
        is_prime = True
        for divisor in range(2, int(candidate ** 0.5) + 1):
            if candidate % divisor == 0:
                is_prime = False
                break
        if is_prime:
            primes.append(candidate)
        candidate += 1

    print(' '.join(map(str, primes)))

# Пример использования:
n = int(input("n = "))
print_first_n_primes(n)
```

```python
# Задание task_02_28.
# Выполнил: Манидарова Н.Ч.
# Группа: ЦИБ-251

#Составьте программу для нахождения всех натуральных решений уравнения x^2+y^2+z^2=k2, где x,y,z∈[1,30], а k вводится с клавиатуры

k=int(input("k = "))

k_squared = k ** 2
solutions = []

for x in range(1, 31):
    for y in range(1, 31):
        for z in range(1, 31):
            if x**2 + y**2 + z**2 == k_squared:
                solutions.append((x, y, z))

if solutions:

    for x, y, z in solutions:
        print(f"x = {x:2d}, y = {y:2d}, z = {z:2d}")
else:
  print(f"Нет решений уравнения")
```

```python
# Задание task_02_29.
# Выполнил: Манидарова Н.Ч.
# Группа: ЦИБ-251

import math

n = int(input("n = "))

nums = []
for i in range(n):
    num = float(input(f"Введите {i+1}-й элемент списка: "))
    nums.append(num)


nums_pos = [x for x in nums if x > 0]


nums_neg = []
for x in nums:
    if x < 0:
        nums_neg.append(x)

# Среднее арифметическое положительных чисел
sr_ar = sum(nums_pos) / len(nums_pos) if nums_pos else 0

if nums_neg:
    product = 1
    for x in nums_neg:
        product *= abs(x)
    sr_geom = product ** (1 / len(nums_neg))
else:
    sr_geom = 0

# Вывод результатов
print(f"Исходный список: {nums}")
print(f"Положительные числа: {nums_pos}")
print(f"Отрицательные числа: {nums_neg}")
print(f"Ср. арифм.: {sr_ar:.2f}")
print(f"Ср. геом.: {sr_geom:.2f}")
```

```python
# Задание task_02_32.
# Выполнил: Манидарова Н.Ч.
# Группа: ЦИБ-251

# 1. Заполнение мест
seats = []
n = int(input("Количество рядов = "))

for i in range(n):
    row_input = input().split()
    row = [int(item) for item in row_input]
    seats.append(row)

# 2. Всего свободных мест
count = 0
for row in seats:
    count += row.count(0)

print(f"Всего свободных мест: {count}")

# 3. Ввод значений для поиска
n_p, m_p = [int(item) for item in
            input("Введите ряд и место через пробел: ").split()]

row_index = n_p - 1
seat_index = m_p - 1

if 0 <= row_index < len(seats) and 0 <= seat_index < len(seats[row_index]):
    is_free = seats[row_index][seat_index] == 0
    print(f"Место свободно: {is_free}")
else:
    print("Такого места не существует")

```

```python
# Задание task_02_33.
# Выполнил: Манидарова Н.Ч.
# Группа: ЦИБ-251


# 1. Заполнение списка

n = int(input("Количество сотрудников = "))
employees = []

for i in range(n):
    emp_data = input(f"{i+1}-й сотрудник (Фамилия Имя Отчество Пол Стаж): ").split()
    # Преобразуем стаж в целое число
    emp_data[4] = int(emp_data[4])
    employees.append(emp_data)

# Вывод списка сотрудников
print("\n".join([" ".join(map(str, emp)) for emp in employees]))

# 2. Самый «молодой» и самый «старый» сотрудник
sorted_employees = sorted(employees, key=lambda x: x[4])
employee_min = sorted_employees[0]
employee_max = sorted_employees[-1]

print(f"Cамый \"молодой\": {' '.join(employee_min[:3])}")
print(f"Cамый \"старый\": {' '.join(employee_max[:3])}")

# 3. Отдельные списки мужчин и женщин
men = [emp for emp in employees if emp[3] == "М"]
women = [emp for emp in employees if emp[3] == "Ж"]

k = input("Введите букву начала имени = ").lower()

# Подсчёт имён, начинающихся на заданную букву
men_k_count = sum(1 for emp in men if emp[1].lower().startswith(k))
women_k_count = sum(1 for emp in women if emp[1].lower().startswith(k))

# Сравнение и вывод результата
if men_k_count > women_k_count:
    print("У мужчин таких имен больше")
elif women_k_count > men_k_count:
    print("У женщин таких имен больше")
else:
    print("Таких имен поровну у мужчин и женщин")
```

```python
# Задание task_02_34.
# Выполнил: Манидарова Н.Ч.
# Группа: ЦИБ-251
# 1. Заполнение списка

n = int(input("Введите количество банков: "))

deposits = []
for _ in range(n):
    data = input().split()
    bank_name = data[0]
    initial_sum = int(data[1])
    rate = float(data[2])

    deposits.append({
        "name": bank_name,
        "initial_sum": initial_sum,
        "rate": rate
    })

# 2. Самый доступный банк (с наименьшей первоначальной суммой)

most_accessible_bank = min(deposits, key=lambda x: x["initial_sum"])

print(f"Самый доступный банк: {most_accessible_bank['name']}")
print(f"Первоначальная сумма: {most_accessible_bank['initial_sum']:.2f} руб.")

# 3. Самый выгодный банк
#    прибыль = сумма * процент / 100

most_profitable_bank = max(
    deposits,
    key=lambda x: x["initial_sum"] * x["rate"] / 100
)

profit = most_profitable_bank["initial_sum"] * most_profitable_bank["rate"] / 100

print(f"Самый выгодный банк: {most_profitable_bank['name']}")
print(f"Прибыль за год: {profit:.2f} руб.")
print(f"Первоначальная сумма: {most_profitable_bank['initial_sum']:.2f} руб.")
print(f"Процентная ставка: {most_profitable_bank['rate']:.2f}%")
```
